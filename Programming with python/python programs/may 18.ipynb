## Problem 2
Generation of sine wave

```python
import numpy as np
t= np.arange(0.0,2.0,0.1)
v=1+np.sin(2*np.pi*t)

plt.xlabel('time')
plt.ylabel('voltage')
plt.grid(True)
plt.plot(t,v)
plt.savefig('figures/time-voltage.png')
plt.show()
```

## problem 3
plot a line graph on income vs year for the state of california

```python
import pandas as pd
df=pd.read_csv('DataFiles/income.csv')
y=df.values[4]
x=df.columns
year =[]
for i in range (2,len(x)):
    year.append(x[i])
year
income=[]
for i in range (2,len(y)):
    income.append(y[i])
income
plt.xlabel('year')
plt.ylabel('income')
plt.title('income of california state')
plt.plot(year,income)
plt.grid(True)
plt.savefig('figures/year-income.png')
plt.show()


    
    
```

```python
width=1/3
plt.bar(year,income,width,color='red')
plt.title('income of california state')
plt.savefig('figures/bar_graph_income_of_california.png')
```

## Problem 4
Generate a scatter plot for 50 random numbers

```python
n=50
x=np.random.rand(n)
y=np.random.rand(n)
colorlist=np.random.rand(n)
arealist=np.pi*(15*np.random.rand(50))**2
plt.scatter(x,y,s=arealist,c=colorlist,alpha=1)
plt.savefig('figures/scatter.png')
plt.show()
```
